```python
import pandas as pd
```

```python
pokemon_df = pd.read_csv('pokemon.csv')
```

```python
pokemon_df.drop(['is_sublegendary','is_legendary','is_mythical'],axis=1, inplace=True)
pokemon_df.rename(columns={'national_number':'Nº'},inplace=True)
```

```python
pokemon_df.head(10)
```

```python
selecao_grama = (pokemon_df['primary_type'] == 'grass') & (pokemon_df['gen'] != 'VIII') & (pokemon_df['gen'] != 'VII') & (pokemon_df['gen'] != 'VI')


pokemon_grama = pokemon_df[selecao_grama]
```

```python
ranking_grama = pokemon_grama[['Nº','gen','Name','primary_type','height_m','weight_kg','hp','attack','defense','sp_attack','sp_defense','speed']]
```

# Top 10 pokemon tipo grama - Melhor HP

```python
ranking_grama.sort_values('hp',ascending=False,ignore_index=True).head(10)
```

# Top 10 pokemon tipo grama - Melhor Ataque

```python
ranking_grama.sort_values('attack',ascending=False,ignore_index=True).head(10)
```
