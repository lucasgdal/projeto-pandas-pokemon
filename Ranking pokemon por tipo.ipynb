```python
import pandas as pd
```

```python
pokemon_df = pd.read_csv('pokemon.csv')
```

```python
pokemon_df.drop(['is_sublegendary','is_legendary','is_mythical'],axis=1, inplace=True)
pokemon_df.rename(columns={'national_number':'Nº'},inplace=True)
```

```python
pokemon_df.head(10)
```

# Ranking dos melhores Pokémon de cada tipo por Atríbuto

```python
tipo = input('Informe o tipo do pokemon (em ingês): ')
selecao_tipo = (pokemon_df['primary_type'] == tipo) & (pokemon_df['gen'] != 'VIII') & (pokemon_df['gen'] != 'VII') & (pokemon_df['gen'] != 'VI')
pokemon_tipo = pokemon_df[selecao_tipo]
```

```python
quantidade_pokemon=pokemon_tipo.shape[0]
print(f'Pokémon tipo: {tipo}')
print(f'Quantidade de Pokémon: {quantidade_pokemon}')
```

```python
ranking_tipo = pokemon_tipo[['Nº','gen','Name','primary_type','height_m','weight_kg','hp','attack','defense','sp_attack','sp_defense','speed']]
```

```python
ranking_tipo.describe().round(2)
```

# Top 10 Pokémon - Melhor HP

```python
ranking_tipo.sort_values('hp',ascending=False,ignore_index=True).head(10)
```
